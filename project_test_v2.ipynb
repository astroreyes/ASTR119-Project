```python
#Import matplotlib and numpy as usual
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## FUNCTION:

#### Our code needs to get any function and run the monte carlo integration through it, so first we need to set our parameters for our function:
#### Note that the bounds for x needs to be between [a,b] and the bounds for y needs to be from [c,d]

```python
# First we will set some parameters of the function
n = 10000 #number of samples for the integration
a = 0.0            #lower x bound
b = (3/2) * np.pi  #upper x bound
c = 0.0            #lower y bound
d = 1.0            #upper y bound
```

```python
#Now we will make a function that will create a function using a provided array 'variable'
def function(variable):
    return np.sin(variable) #RETURNS FUNCTION
```

## MONTE CARLO INTEGRATION METHOD:

### For our Monte Carlo Integration, we will create uniformly sampled variables that will integrate over our function

```python
#First we will make uniformly sampled variables
# Our sample has to be slightly bigger than that of the bounds of the given f(x)
def monte_carlo_parameters(lower_x,upper_x,lower_y,upper_y):
    a = lower_x - 0.1
    b = upper_x + 0.1
    c = lower_y - 0.1
    d = upper_y + 0.1
    return a,b,c,d

samples = monte_carlo_parameters(a,b,c,d)

#Now we can make our random samples:
x = np.random.uniform(samples[0],samples[1],n) #creates an array of uniformly random samples 
y = np.random.uniform(-1.1,1.1,n)              #creates an array of uniformly random samples 

#Now we create an array and make a function f(x)
var = np.linspace(a,b,n) # makes an array [a,b] with n samples
f_x = function(var)      # makes a function f(x)

#Now we can check to see if our samples are within our bounds:
ir = np.where((y)<f_x)[0] # gives samples below f(x)
ur = np.where((y)>f_x)[0] # gives samples above f(x) 
```

## GRAPH:

```python
fig = plt.figure(figsize=(7,7))
plt.xlim([samples[0],samples[1]]) #This will confine my graph from [-0.1,1.1] in x-axis
plt.ylim([-1.1,1.1]) #This will confine my graph from [-0.1,1.1] in y-axis
plt.plot(x[ir],y[ir],'.',color='blue') #This plots all my sample points
plt.plot(x[ur],y[ur],'.',color='0.75') #This plots points below the curve
plt.plot(var,f_x,color='green',linewidth=4.0) #Plots the given function
plt.hlines(0,-1,5,"red",linewidth=4.0) #Plots the x-axis

plt.show()
```
