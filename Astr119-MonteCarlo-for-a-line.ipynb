```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np
```

## Parameters of integration:

```python
n = 1000 #this is the number of samples that we are looking at for our integration
```

## Uniformally sampled variables for a line:

where we are finding the integral of y=1/2

```python
x = np.random.uniform(0,.5,n) 
y = np.random.uniform(0,1,n)
```

```python
ir = np.where((y)<.5)[0] #inside
ur = np.where((y)>=.5)[0] #outside
```

## plot the samples and our line... (hopefully)

```python
fig = plt.figure(figsize=(7,7))
plt.xlim([-.1,.6])
plt.ylim([-.1,1.1])
plt.plot(x[ir],y[ir],'.',color="blue")
plt.plot(x[ur],y[ur],'.',color="0.75")

#plotting the line

z = np.linspace(-1,1,1000)
xl = z
yl = z*0+.5

plt.plot(xl,yl,color="green")

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
