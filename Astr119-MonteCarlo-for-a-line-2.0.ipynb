```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np
```

## Parameters of integration:

```python
n = 10000 #this is the number of samples that we are looking at for our integration 
#ty aj for noticing that I didn't have enough points
```

## Uniformally sampled variables for a line:

where we are finding the integral of y=1/2

```python
x = np.random.uniform(-1,1,n) 
y = np.random.uniform(-1,1,n)
```

```python
ir = np.where((y)<.5) [0] #inside
ur = np.where((y)>=.5)[0] #outside
br = np.where((y)<=0) [0] #bottom line this is important lol 
```

## plot the samples and our line... (hopefully)

```python
fig = plt.figure(figsize=(7,7))
plt.xlim([-1.1,1.1])
plt.ylim([-1.1,1.1])
plt.plot(x[ir],y[ir],'.',color="blue")
plt.plot(x[ur],y[ur],'.',color="0.75")
plt.plot(x[br],y[br],'.',color="0.75")

#plotting the line

z = np.linspace(-1,1,10000)
xl = z
yl = z*0+.5


plt.plot(xl,yl,color="green")

#plottting the bottom line 

b= np.linspace(-1,1,10000)
xb = z
yb = z*0

plt.plot(xb,yb,color="green")

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
